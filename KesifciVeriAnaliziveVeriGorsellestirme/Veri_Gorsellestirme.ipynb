```python
import seaborn as sns
diamonds =sns.load_dataset("diamonds")
df = diamonds.copy()
df.head()
```

```python
# ordinal tanımlama
from pandas.api.types import CategoricalDtype
```

```python
df.cut.head()
```

```python
df.cut.astype(CategoricalDtype (ordered =True))
```

#### burada cut sınıflarının verilişi yanlış bunları doğru sıralamada yani veri setinin hikayeisine göre sıralamak için aşağıdaki yöntem kullanılır.

```python
cut_categories = ["Fair", "Good", "Very Good", "Premium", "Ideal"]
```

```python
df.cut.astype(CategoricalDtype (categories = cut_categories ,ordered =True)) # burada argüman olarak doğru sıralamayı veriyoruz.
```

## Sütun grafiklerini oluşturulması "barplot"

```python
df["cut"].value_counts().plot.barh();
```

```python
sns.barplot(x = "cut", y= df.cut.index, data = df);
```

### Çaprazlamalar

```python
sns.catplot(x = "cut", y= "price", data = df);
```

```python
sns.barplot(x="cut", y="price", hue="color", data= df);
```

```python
df.groupby(["cut", "color"])["price"].mean()
```

### Histogram ve Yoğunluk

```python
sns.histplot(df.price,  kde= False);
```

```python
sns.histplot(df.price, kde=True);
```

```python
sns.kdeplot(df.price, fill=True);
```

### Histogram ve Yoğunluk Caprazlamalar

```python
(sns
.FacetGrid(df,
              hue = "cut",
              height = 5,
              xlim = (0, 10000))
.map(sns.kdeplot, "price", fill = True)
.add_legend()
);
```

```python
sns.catplot(x="cut", y="price", hue="color", kind="point", data= df);
```
