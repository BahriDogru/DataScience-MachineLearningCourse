### Isı Haritası (Heatmap)

```python
flights = sns.load_dataset("flights")
df2 = flights.copy()
df2.head()
```

```python
df2.shape
```

```python
df2 = df2.pivot(index="year", columns="month", values="passengers");
```

```python
df2
```

```python
sns.heatmap(df2);
```

```python
sns.heatmap(df2, annot=True, fmt ="d");
```

```python
sns.heatmap(df2, annot=True, fmt ="d", linewidths = 0.5);
```

 ### Çizgi Grafik 

```python
fmri = sns.load_dataset("fmri")
df = fmri.copy()
df.head()
```

Her bir timepoint'e göre signal'ı gözlemleme

```python
df["timepoint"].describe()
```

```python
df["signal"].describe()
```

```python
df.groupby("timepoint")["signal"].count()
```

```python
df.groupby("signal").count()
```

```python
df.groupby("timepoint")["signal"].describe()
```

### Çizgi Grafiği ve Çaprazlamalar

```python
sns.lineplot(x ="timepoint", y = "signal", data = df);
```

```python
sns.lineplot(x = "timepoint", y = "signal", hue = "event", data = df); 
```

```python
sns.lineplot(x = "timepoint", y = "signal", hue = "event", style = "event", data = df); 
```

```python
sns.lineplot(x = "timepoint",
             y = "signal",
             hue = "event",
             style = "event",
             markers = True,
             dashes = False,
             data = df); 
```

```python
sns.lineplot(x = "timepoint",
             y = "signal",
             hue = "region",
             style = "event",
             data = df); 
```
